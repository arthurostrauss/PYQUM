```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_experiments.library.randomized_benchmarking import InterleavedRB
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit.circuit.library import CZGate, XGate, HGate, RXGate
import numpy as np
# Create a noise model from depolarizing error
noise_model = NoiseModel()
X90 = RXGate(np.pi/2)
p1Q = 0.05  # 1-qubit depolarizing error probability
p2Q = 0.01   # 2-qubit depolarizing error probability
noise_model.add_all_qubit_quantum_error(depolarizing_error(p1Q, 1), ['rx'])
# noise_model.add_all_qubit_quantum_error(depolarizing_error(p1Q, 1), 'u2')
# noise_model.add_all_qubit_quantum_error(depolarizing_error(p2Q, 2), 'cz')

backend = AerSimulator(noise_model=noise_model, basis_gates=noise_model.basis_gates)
# backend = FakeJakartaV2()
# Create a 2-qubit interleaved RB experiment
rb = InterleavedRB(X90, [0,], lengths=[1, 10, 20, 40, 50, 75, 100], backend=backend)
results = rb.run()
print(results)
```
